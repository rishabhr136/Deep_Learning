```python
df_train = pd.read_csv("/content/drive/MyDrive/df_processed_train.csv")
df_train.head()
```

```python
x_train = df_train.iloc[:,1:-1]
y_train = df_train.iloc[:,-1]
```

```python
x_train.head()
```

```python
df_val =  pd.read_csv("/content/drive/MyDrive/df_processed_val.csv")
```

```python
df_val.head()
```

```python
x_val = df_val.iloc[:,1:-1]
y_val = df_val.iloc[:,-1]
x_val.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense
#Design the deep neural network [Small + 1 layer]
model = Sequential()
model.add(Dense(256,input_dim = x_train.shape[1],activation= "relu"))
model.add(Dense(256,activation="relu"))
model.add(Dense(1,activation = "sigmoid")) #activation = sigmoid for binary classification
model.compile(optimizer = "Adam",loss="binary_crossentropy", metrics=["accuracy"])
model.fit(x_train,y_train, validation_data = (x_val,y_val), epochs=3, batch_size=64)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_val_scaled = scaler.transform(x_val)
```
